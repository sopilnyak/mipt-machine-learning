```python
%matplotlib inline

import numpy as np
import pylab as plt
'''
Performs the Principal Coponent analysis of the Matrix F
Matrix must be n * l dimensions
where n is # features
l is # samples
'''

def PCA(F, varRetained = 0.95, show = False):
    # Input
    #     F - initaial matrix 
    # Compute Covariance Matrix Sigma
    # Input
    (n, l) = F.shape
    Sigma = 1.0 / l * np.dot(F, np.transpose(F))
    # Compute eigenvectors and eigenvalues of Sigma by SVD
    # U, V - matrix, d - array: Sigma = U * np.diag(d) * V
    U, d, V = np.linalg.svd(Sigma)

    # compute the value m: number of minumum features that retains the given variance varRetained
    dTot = np.sum(d)
    var_i = np.array([np.sum(d[: i + 1]) / \
                dTot * 100.0 for i in range(n)])
    m = len(var_i[var_i > (varRetained * 100.0)])  # необходимое число главных компонент
    print '%.2f %% variance retained in %d dimensions' % (var_i[m], m)

    # plot the variance plot
    if show:
        plt.plot(var_i)
        plt.xlabel('Number of Features')
        plt.ylabel(' Percentage Variance retained')
        plt.title('PCA $\% \sigma^2 $ vs # features')
        plt.show()

    # compute the reduced dimensional features by projection
    U_reduced = U[:, :m]  #только m главных компонент
    G = np.dot(np.transpose(U_reduced), F)

    return G, U_reduced
```

```python
from sklearn.datasets import load_digits
from sklearn.cluster import DBSCAN
from sklearn import cluster, metrics
from sklearn.decomposition import PCA
```

```python
mnist = load_digits()
digits = np.reshape(mnist.images, (mnist.images.shape[0], -1))
```

Оценим качество различных кластеризаций при помощи **Adjusted Mutual Information** и **Adjusted Rand Index** для разных значений параметров eps и min_samples.

```python
best_eps1 = 0
best_min_samples1 = 0
best_eps2 = 0
best_min_samples2 = 0
adjusted_rand_score = 0
adjusted_mutual_info_score = 0
```

```python
for eps in np.linspace(3.5, 30., 100):
    
    for min_samples in np.arange(3, 20):
        
        labels = DBSCAN(eps=eps, min_samples=min_samples).fit_predict(digits)
        if metrics.adjusted_mutual_info_score(labels, mnist.target) > adjusted_mutual_info_score:
            best_eps1 = eps
            best_min_samples1 = min_samples
            
        if metrics.adjusted_rand_score(labels, mnist.target) > adjusted_rand_score:
            best_eps2 = eps
            best_min_samples2 = min_samples
```

Лучшее значение eps получилось около 21, поэтому попробуем применить его.

```python
#digits_pca = PCA(n_components=10).fit_transform(digits)

dbscan = DBSCAN(eps=21)  # по умолчанию метрика euclidean
labels = dbscan.fit_predict(digits)

metrics.adjusted_mutual_info_score(labels, mnist.target)
```

```python
metrics.adjusted_rand_score(labels, mnist.target)
```
